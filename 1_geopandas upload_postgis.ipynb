```python
from sqlalchemy import create_engine, MetaData
import geopandas as gpd

## Connect to the Database 
conf ={
  'host':"x",
  'port':'x',
  'database':"x",
  'user':"x",
  'password':"x"
}
# Run Engine
engine = create_engine("postgresql://{user}:{password}@{host}:{port}/{database}".format(**conf))

# Reflect the database tables
metadata = MetaData()
metadata.reflect(bind=engine)
# Print the table names
print("Existing tables:")
for table in metadata.tables.values():
  print(table.name)
```

```python
# Deprecated
import geopandas as gpd
from sqlalchemy import create_engine
import pandas as pd
import gc

# Function to read data from PostGIS in chunks
def read_postgis_chunks(table_name, db_connection_str, chunksize=1000):
    engine = create_engine(db_connection_str)
    chunks = pd.read_sql_table(table_name, con=engine, chunksize=chunksize)
    return chunks

# Function to write data to PostGIS in chunks
def write_postgis_chunks(gdf, table_name, db_connection_str, chunksize=1000):
    engine = create_engine(db_connection_str)
    for i, chunk in enumerate(gdf.groupby(gdf.index // chunksize)):
        chunk.to_postgis(table_name, engine, if_exists='append', index=False, chunksize=chunksize)
        # Clean up memory after each chunk
        del chunk
        gc.collect()

# Example usage
gdb_path = r"C:\Users\guilh\Downloads\car.gdb"
table_name = 'car'

# Step 1: Read the GDB file in chunks using GeoPandas
chunks_reader = gpd.read_file(gdb_path, driver='FileGDB', chunksize=1000)

# Step 2: Write each chunk to PostGIS
for chunk in chunks_reader:
    write_postgis_chunks(chunk, table_name, engine)
    # Clean up memory after writing each chunk
    del chunk
    gc.collect()

print("Data successfully added to PostGIS.")
```
